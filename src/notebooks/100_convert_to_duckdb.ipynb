```python
from pathlib import Path

import duckdb
import pandas as pd
```

```python
DATA_DIR = Path.cwd().parent.parent / "data"

# Read in data
df = pd.read_excel(DATA_DIR / "LIHTCPUB.xlsx")

# Connect to duckdb
conn = duckdb.connect(str(DATA_DIR / "lihtc.duckdb"))

# Copy data from the dataframe
conn.execute("CREATE TABLE lihtc AS SELECT * FROM df;")

# Make sure it worked
conn.query("SELECT * FROM lihtc LIMIT 10").to_df()
```

```python
# Add the other database as well
df = pd.read_excel(DATA_DIR / "LIHTCPUB_BIN.xlsx")

# Copy data from the dataframe
conn.execute("CREATE TABLE lihtc_bin AS SELECT * FROM df;")

# Make sure it worked
conn.query("SELECT * FROM lihtc_bin LIMIT 10").to_df()
```

```python
conn.close()
```
